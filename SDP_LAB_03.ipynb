## **Builder Pattern Apply the to model a computer system where different configurations can be created by selecting components such as CPU, GPU, internal and external memory, operating system, touchscreen capability, Bluetooth mouse, and Bluetooth keyboard.**

### **Product Class**

```python
class Computer:
    def __init__(self):
        self.cpu = None
        self.gpu = None
        self.internal_memory = None
        self.external_memory = None
        self.operating_system = None
        self.touchscreen = False
        self.bluetooth_mouse = False
        self.bluetooth_keyboard = False

    def display_configuration(self):
        print("Computer Configuration:")
        print(f"CPU               : {self.cpu}")
        print(f"GPU               : {self.gpu}")
        print(f"Internal Memory   : {self.internal_memory}")
        print(f"External Memory   : {self.external_memory}")
        print(f"Operating System  : {self.operating_system}")
        print(f"Touchscreen       : {'Yes' if self.touchscreen else 'No'}")
        print(f"Bluetooth Mouse   : {'Yes' if self.bluetooth_mouse else 'No'}")
        print(f"Bluetooth Keyboard: {'Yes' if self.bluetooth_keyboard else 'No'}")
```

### **Builder Class**

```python
class ComputerBuilder:
    def __init__(self):
        self.computer = Computer()

    def set_cpu(self, cpu):
        self.computer.cpu = cpu
        return self

    def set_gpu(self, gpu):
        self.computer.gpu = gpu
        return self

    def set_internal_memory(self, memory):
        self.computer.internal_memory = memory
        return self

    def set_external_memory(self, memory):
        self.computer.external_memory = memory
        return self

    def set_operating_system(self, os):
        self.computer.operating_system = os
        return self

    def enable_touchscreen(self):
        self.computer.touchscreen = True
        return self

    def add_bluetooth_mouse(self):
        self.computer.bluetooth_mouse = True
        return self

    def add_bluetooth_keyboard(self):
        self.computer.bluetooth_keyboard = True
        return self

    def build(self):
        return self.computer
```

### **Create Different Configurations**

**Gaming Computer**

```python
gaming_pc = (ComputerBuilder()
             .set_cpu("Intel Core i9")
             .set_gpu("NVIDIA RTX 4090")
             .set_internal_memory("32 GB RAM")
             .set_external_memory("2 TB SSD")
             .set_operating_system("Windows 11")
             .add_bluetooth_mouse()
             .add_bluetooth_keyboard()
             .build())

gaming_pc.display_configuration()
```

**Office Laptop**

```python
office_laptop = (ComputerBuilder()
                 .set_cpu("Intel Core i5")
                 .set_gpu("Integrated Graphics")
                 .set_internal_memory("16 GB RAM")
                 .set_external_memory("512 GB SSD")
                 .set_operating_system("Ubuntu Linux")
                 .enable_touchscreen()
                 .add_bluetooth_mouse()
                 .build())

office_laptop.display_configuration()
```
