### **1) Employee Salary Calculation**

```python
from abc import ABC, abstractmethod

# Abstract Base Class
class Employee(ABC):
    def __init__(self, base_salary, bonus):
        self.base_salary = base_salary
        self.bonus = bonus

    @abstractmethod
    def calculate_salary(self):
        pass
```

**Permanent Employee**

```python
class PermanentEmployee(Employee):
    def __init__(self, base_salary, bonus, hra):
        super().__init__(base_salary, bonus)
        self.hra = hra

    def calculate_salary(self):
        return self.base_salary + self.bonus + self.hra
```

**Contract Employee**

```python
class ContractEmployee(Employee):
    def __init__(self, base_salary, bonus, medical_allowance):
        super().__init__(base_salary, bonus)
        self.medical_allowance = medical_allowance

    def calculate_salary(self):
        return self.base_salary + self.bonus + self.medical_allowance
```

**Employee Factory (No if-else )**

```python
class EmployeeFactory:
    _creators = {}

    @classmethod
    def register(cls, emp_type, creator):
        cls._creators[emp_type] = creator

    @classmethod
    def create(cls, emp_type, *args):
        creator = cls._creators.get(emp_type)
        if not creator:
            raise ValueError(f"Unknown employee type: {emp_type}")
        return creator(*args)
```

**Register Types**

```python
EmployeeFactory.register("permanent", PermanentEmployee)
EmployeeFactory.register("contract", ContractEmployee)
```

**Usage**

```python
emp1 = EmployeeFactory.create("permanent", 50000, 5000, 10000)
emp2 = EmployeeFactory.create("contract", 40000, 3000, 5000)

print("Permanent Employee Salary:", emp1.calculate_salary())
print("Contract Employee Salary:", emp2.calculate_salary())
```

### **2) Bank Account Balance Calculation**

**Base Class**

```python
class BankAccount(ABC):
    def __init__(self, balance):
        self.balance = balance

    @abstractmethod
    def calculate_balance(self):
        pass
```

**Savings Account**

```python
class SavingsAccount(BankAccount):
    def __init__(self, balance, interest_rate):
        super().__init__(balance)
        self.interest_rate = interest_rate

    def calculate_balance(self):
        return self.balance + (self.balance * self.interest_rate / 100)
```

**Current Account**

```python
class CurrentAccount(BankAccount):
    def __init__(self, balance, maintenance_fee):
        super().__init__(balance)
        self.maintenance_fee = maintenance_fee

    def calculate_balance(self):
        return self.balance - self.maintenance_fee
```

**Account Factory**

```python
class AccountFactory:
    _creators = {}

    @classmethod
    def register(cls, acc_type, creator):
        cls._creators[acc_type] = creator

    @classmethod
    def create(cls, acc_type, *args):
        creator = cls._creators.get(acc_type)
        if not creator:
            raise ValueError(f"Unknown account type: {acc_type}")
        return creator(*args)
```

**Register Types**

```python
AccountFactory.register("savings", SavingsAccount)
AccountFactory.register("current", CurrentAccount)
```

**Usage**

```python
acc1 = AccountFactory.create("savings", 10000, 5)
acc2 = AccountFactory.create("current", 10000, 500)

print("Savings Account Balance:", acc1.calculate_balance())
print("Current Account Balance:", acc2.calculate_balance())
```
