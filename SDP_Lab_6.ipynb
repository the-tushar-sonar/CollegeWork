## **1. Decorator Pattern**
In a slot machine, we want to add a temporary 2x bonus multiplier that expires after 5 spins. How would you modify the MultiplierDecorator class to implement this?

  The solution should:Apply the multiplier only for the first 5 spins.

  Automatically stop applying it afterward.

  Not break existing payout calculations when inactive.

```python
# Base Component
class SlotMachine:
    def spin(self):
        return 100  # base payout


# Decorator Base Class
class SlotMachineDecorator(SlotMachine):
    def __init__(self, slot_machine):
        self.slot_machine = slot_machine

    def spin(self):
        return self.slot_machine.spin()


# Multiplier Decorator with 5-spin limit
class MultiplierDecorator(SlotMachineDecorator):
    def __init__(self, slot_machine, multiplier=2, max_spins=5):
        super().__init__(slot_machine)
        self.multiplier = multiplier
        self.remaining_spins = max_spins

    def spin(self):
        base_payout = self.slot_machine.spin()

        if self.remaining_spins > 0:
            self.remaining_spins -= 1
            print(f"Bonus Applied! Spins left: {self.remaining_spins}")
            return base_payout * self.multiplier
        else:
            print("No bonus applied.")
            return base_payout


# Testing Decorator
print("Decorator Pattern Output:")
machine = SlotMachine()
bonus_machine = MultiplierDecorator(machine)

for i in range(7):  # 7 spins
    print(f"Spin {i+1} payout:", bonus_machine.spin())
```

## **2. Proxy Pattern Question**
"Implement a SlotMachineProxy that blocks spins during maintenance hours (e.g., 2 AM to 6 AM).
The solution should:

Reject spins with a "Under maintenance" message during the blocked time window.

Allow normal spins outside maintenance hours.

Not require changes to the real SlotMachine class.


```python
from datetime import datetime

# Real Subject
class SlotMachine:
    def spin(self):
        return "Spinning... You won 100 coins!"


# Proxy
class SlotMachineProxy:
    def __init__(self, real_machine):
        self.real_machine = real_machine

    def spin(self):
        current_hour = datetime.now().hour

        # Maintenance time: 2 AM to 6 AM
        if 2 <= current_hour < 6:
            return "⚠️ Under maintenance. Try later."
        else:
            return self.real_machine.spin()


# Testing Proxy
print("\nProxy Pattern Output:")
real_machine = SlotMachine()
proxy = SlotMachineProxy(real_machine)

print("Current Time:", datetime.now().strftime("%H:%M:%S"))
print(proxy.spin())
```
