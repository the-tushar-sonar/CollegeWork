## **1. Imagine you are operating an e-commerce system where orders pass through several stages: (Use chain of responsibility)**

* Validation: To ensure the order details are complete and accurate.
* Discount Application: To check and apply any available discounts or promotional offers.
* Payment: To process payment for the order.
* Shipping: To finally ship the order to the customer.

```python
# Base Handler
class Handler:
    def __init__(self):
        self.next_handler = None

    def set_next(self, handler):
        self.next_handler = handler
        return handler

    def handle(self, order):
        if self.next_handler:
            return self.next_handler.handle(order)


# Concrete Handlers
class ValidationHandler(Handler):
    def handle(self, order):
        if not order.get("valid"):
            return "Order validation failed"
        print("Validation done")
        return super().handle(order)


class DiscountHandler(Handler):
    def handle(self, order):
        if order.get("discount"):
            print("Discount applied")
        return super().handle(order)


class PaymentHandler(Handler):
    def handle(self, order):
        print("Payment processed")
        return super().handle(order)


class ShippingHandler(Handler):
    def handle(self, order):
        print("Order shipped")
        return "Order completed"


# Setup chain
validation = ValidationHandler()
discount = DiscountHandler()
payment = PaymentHandler()
shipping = ShippingHandler()

validation.set_next(discount).set_next(payment).set_next(shipping)

# Test
order = {"valid": True, "discount": True}
print("Chain Output:")
print(validation.handle(order))
```

## **2. Imagine you have a bookshelf, and you want to provide a way for readers to browse books on the shelf without knowing how the books are organized. (Use Iterator)**

```python
class BookShelf:
    def __init__(self):
        self.books = []

    def add_book(self, book):
        self.books.append(book)

    def __iter__(self):
        return BookIterator(self.books)


class BookIterator:
    def __init__(self, books):
        self.books = books
        self.index = 0

    def __next__(self):
        if self.index < len(self.books):
            book = self.books[self.index]
            self.index += 1
            return book
        raise StopIteration


# Test
shelf = BookShelf()
shelf.add_book("DSA")
shelf.add_book("OS")
shelf.add_book("DBMS")

print("\nIterator Output:")
for book in shelf:
    print(book)
```


## **3. Consider a smart home system where different components like lights, windows, curtains, and heaters can interact with each other through a central hub or controller. (Use Mediator)**

```python
class Mediator:
    def notify(self, sender, event):
        pass


class SmartHomeMediator(Mediator):
    def __init__(self):
        self.devices = {}

    def register(self, name, device):
        self.devices[name] = device
        device.mediator = self

    def notify(self, sender, event):
        if event == "motion_detected":
            print("Lights ON, Curtains OPEN")
        elif event == "cold":
            print("Heater ON")


class Device:
    def __init__(self, mediator=None):
        self.mediator = mediator


class Sensor(Device):
    def detect_motion(self):
        print("Motion detected")
        self.mediator.notify(self, "motion_detected")


# Test
mediator = SmartHomeMediator()
sensor = Sensor()

mediator.register("sensor", sensor)

print("\nMediator Output:")
sensor.detect_motion()
```

## **4. A social media platform where users can follow other users and get notified when someone they follow posts new content. Imagine a platform similar to Twitter where users can post messages (tweets), and followers get notified of these messages. (Use Observer)**

```python
# Subject
class User:
    def __init__(self, name):
        self.name = name
        self.followers = []

    def follow(self, follower):
        self.followers.append(follower)

    def post(self, message):
        print(f"{self.name} posted: {message}")
        for follower in self.followers:
            follower.update(self.name, message)


# Observer
class Follower:
    def __init__(self, name):
        self.name = name

    def update(self, user, message):
        print(f"{self.name} got notified: {user} posted '{message}'")


# Test
user1 = User("Piyush")
f1 = Follower("Ajay")
f2 = Follower("Kalpesh")

user1.follow(f1)
user1.follow(f2)

print("\nObserver Output:")
user1.post("Hello World!")
```
