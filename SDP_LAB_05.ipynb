## **1: Bridge Design Pattern (Payment Gateway)**

### **Implementation**
**Step 1: Implementor (Payment Processor)**

```python
from abc import ABC, abstractmethod

class PaymentProcessor(ABC):
    @abstractmethod
    def process_payment(self, amount):
        pass
```

**Concrete Implementations**

```python
class StripeProcessor(PaymentProcessor):
    def process_payment(self, amount):
        return f"Processed ₹{amount} via Stripe"

class SquareProcessor(PaymentProcessor):
    def process_payment(self, amount):
        return f"Processed ₹{amount} via Square"

class PayPalProcessor(PaymentProcessor):
    def process_payment(self, amount):
        return f"Processed ₹{amount} via PayPal API"
```

**Step 2: Abstraction (Payment Method)**

```python
class PaymentMethod(ABC):
    def __init__(self, processor):
        self.processor = processor

    @abstractmethod
    def pay(self, amount):
        pass
```

**Refined Abstractions**

```python
class CreditCard(PaymentMethod):
    def pay(self, amount):
        return f"Credit Card → {self.processor.process_payment(amount)}"

class PayPal(PaymentMethod):
    def pay(self, amount):
        return f"PayPal → {self.processor.process_payment(amount)}"

class Crypto(PaymentMethod):
    def pay(self, amount):
        return f"Crypto → {self.processor.process_payment(amount)}"
```

**Usage**

```python
processor = StripeProcessor()
payment = CreditCard(processor)
print(payment.pay(5000))

payment2 = Crypto(SquareProcessor())
print(payment2.pay(7000))
```

## **2: Composite Pattern (University – Iterator)**

**Implementation**

```python
class Student:
    def __init__(self, name):
        self.name = name

    def __str__(self):
        return self.name
```

**Course (Composite)**

```python
class Course:
    def __init__(self, name):
        self.name = name
        self.students = []

    def add_student(self, student):
        self.students.append(student)

    def __iter__(self):
        return iter(self.students)
```

**University (Iterator over all courses)**

```python
class University:
    def __init__(self):
        self.courses = []

    def add_course(self, course):
        self.courses.append(course)

    def get_all_students(self):
        for course in self.courses:
            for student in course:
                yield student
```

**Usage**

```python
c1 = Course("CS")
c1.add_student(Student("Harshal"))
c1.add_student(Student("Kalpesh"))

c2 = Course("IT")
c2.add_student(Student("Humanshu"))

uni = University()
uni.add_course(c1)
uni.add_course(c2)

print("All Students:")
for s in uni.get_all_students():
    print(s)
```

## **3: Composite Pattern (Company Structure – Based on Image)**

### **Implementation**
**Component**

```python
from abc import ABC, abstractmethod

class EmployeeComponent(ABC):
    @abstractmethod
    def show_details(self):
        pass

    @abstractmethod
    def get_salary(self):
        pass
```

**Leaf (Employee)**

```python
class Employee(EmployeeComponent):
    def __init__(self, name, salary):
        self.name = name
        self.salary = salary

    def show_details(self):
        print(f"Name: {self.name} and Salary: {self.salary}")

    def get_salary(self):
        return self.salary
```

**Composite (Department / Company)**

```python
class CompositeEmployee(EmployeeComponent):
    def __init__(self, name):
        self.name = name
        self.employees = []

    def add(self, emp):
        self.employees.append(emp)

    def show_details(self):
        for emp in self.employees:
            emp.show_details()

    def get_salary(self):
        return sum(emp.get_salary() for emp in self.employees)
```

**Usage (Matches Your Output)**

```python
# Employees
e1 = Employee("Pranaya", 22000)
e2 = Employee("Rohit", 33000)
e3 = Employee("Anurag", 42000)
e4 = Employee("Priyanka", 52000)
e5 = Employee("Sambit", 28000)

# Departments
it_dept = CompositeEmployee("IT Department")
it_dept.add(e1)
it_dept.add(e2)
it_dept.add(e3)

hr_dept = CompositeEmployee("HR Department")
hr_dept.add(e4)
hr_dept.add(e5)

# Company
company = CompositeEmployee("Company")
company.add(it_dept)
company.add(hr_dept)

# Output
print("All Employees of the Company")
company.show_details()
print("Total Salary of the Company:", company.get_salary())

print("\nIT Department Employees")
it_dept.show_details()
print("Total Salary of IT Department:", it_dept.get_salary())

print("\nHR Department Employees")
hr_dept.show_details()
print("Total Salary of HR Department:", hr_dept.get_salary())

print("\nSalary of Employee Pranaya:")
e1.show_details()

print("\nSalary of Employee Anurag:")
e3.show_details()
```
